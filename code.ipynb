```python
import mysql.connector
from mysql.connector import Error
import tkinter as tk
from tkinter import messagebox, ttk

def create_connection():
    connection = None
    try:
        connection = mysql.connector.connect(
            host="localhost",
            user="root",
            password="Mohit@123",
            database="ExpenseTrac"
             )
        if connection.is_connected():
            print("Successfully connected to the database")
    except Error as e:
        print(f"The error '{e}' occurred")
    return connection
def add_expense(connection, amount, category, description):
    cursor = connection.cursor()
    query = "INSERT INTO expenses (amount, category, description) VALUES (%s, %s, %s)"
    cursor.execute(query, (amount, category, description))
    connection.commit()
    
def delete_expense(connection, expense_id):
    cursor = connection.cursor()
    query = "DELETE FROM expenses WHERE id = %s"
    cursor.execute(query, (expense_id,))
    connection.commit()
def update_expense(connection, expense_id, amount, category, description):
    cursor = connection.cursor()
    query = "UPDATE expenses SET amount = %s, category = %s, description = %s WHERE id = %s"
    cursor.execute(query, (amount, category, description, expense_id))
    connection.commit()
def get_expenses(connection):
    cursor = connection.cursor()
    query = "SELECT * FROM expenses"
    cursor.execute(query)
    return cursor.fetchall()
class ExpenseTrac:
    def __init__(self, root):
        self.connection = create_connection()
        self.root = root
        self.root.title("Expense Trac")
        
        self.create_widgets()
        
    def create_widgets(self):
        self.amount_label = tk.Label(self.root, text="Amount:")
        self.amount_label.grid(row=1, column=0)
        self.amount_entry = tk.Entry(self.root)
        self.amount_entry.grid(row=1, column=1)
        
        self.category_label = tk.Label(self.root, text="Category:")
        self.category_label.grid(row=2, column=0)
        self.category_entry = tk.Entry(self.root)
        self.category_entry.grid(row=2, column=1)
        
        self.description_label = tk.Label(self.root, text="Description:")
        self.description_label.grid(row=3, column=0)
        self.description_entry = tk.Entry(self.root)
        self.description_entry.grid(row=3, column=1)

        self.add_button = tk.Button(self.root, text="Add Expense", command=self.add_expense)
        self.add_button.grid(row=4, column=0, columnspan=2)
        
        self.update_button = tk.Button(self.root, text="Update Expense", command=self.update_expense)
        self.update_button.grid(row=5, column=0, columnspan=2)
        
        self.delete_button = tk.Button(self.root, text="Delete Expense", command=self.delete_expense)
        self.delete_button.grid(row=6, column=0, columnspan=2)

        self.expenses_list = ttk.Treeview(self.root, columns=("ID", "Amount", "Category", "Description"), show='headings')
        self.expenses_list.heading("ID", text="ID")
        self.expenses_list.heading("Amount", text="Amount")
        self.expenses_list.heading("Category", text="Category")
        self.expenses_list.heading("Description", text="Description")
        self.expenses_list.grid(row=7, column=0, columnspan=2)
        
        self.load_expenses()
 
    def add_expense(self):
        amount = self.amount_entry.get()
        category = self.category_entry.get()
        description = self.description_entry.get()
        add_expense(self.connection, amount, category, description)
    
        self.load_expenses()

    def update_expense(self):
        selected_item = self.expenses_list.selection()[0]
        expense_id = self.expenses_list.item(selected_item)['values'][0]
        amount = self.amount_entry.get()
        category = self.category_entry.get()
        description = self.description_entry.get()
        update_expense(self.connection, expense_id, amount, category, description)
        self.load_expenses()

   
    def delete_expense(self):
        selected_item = self.expenses_list.selection()[0]
        expense_id = self.expenses_list.item(selected_item)['values'][0]
        delete_expense(self.connection, expense_id)
        self.load_expenses()

    
    def load_expenses(self):
        for row in self.expenses_list.get_children():
            self.expenses_list.delete(row)
        expenses = get_expenses(self.connection)
        for expense in expenses:
            self.expenses_list.insert('', 'end', values=expense)

if __name__ == "__main__":
    root = tk.Tk()
    app = ExpenseTrac(root)
    root.mainloop()
```
